_Now, let's find a way to programmatically clean this data up_
### Let's use the helpful `transform` method!
_Transform works similarly to `apply`, where you pass a function to it and iteravely applies it to columns or rows, and it's way way faster than using `.iterrows` much less a dreaded `for loop` which we should always try to avoid with pandas if we can!_

````python
# help(df_json_unclean.transform)
````

### Let's first convert the "count" columns to integers directly in the series

````python
def count_to_int(col):
    return int(col['count'])
````

````python
df_raw_social_data['likes'] = df_raw_social_data['edge_media_preview_like'].transform(count_to_int)
df_raw_social_data['n_comments'] = df_raw_social_data['edge_media_to_comment'].transform(count_to_int)
````

_Let's look at how we did._

````python
df_raw_social_data[['likes', 'n_comments']].head()
````

_Yay! Seems to have worked!_

#### Let's burrow into those caption dictionaries.

````python
# Yikes! 
# I did it ahead of time for this one, but basically same process as before. 
df_raw_social_data['edge_media_to_caption'][0]['edges'][0]['node']['text']
````

#### Besides being faster than a straight for loop and other iteration-type methods, one cool thing about applying functions this way in Pandas is that you can handle errors in the data gracefully, then get a sense for if it's OK or not.

````python
def caption_dict_to_string(col):
    try:
        return col['edges'][0]['node']['text']
    except Exception as e:
        print("Data was corrupted for a caption")
        return "No caption downloaded or caption was corrupted"
````

````python
df_raw_social_data['caption'] = df_raw_social_data['edge_media_to_caption'].apply(caption_dict_to_string)
````

## üòµ 

### Let's see if I can simplify it a little bit for us.

`.rolling()` basically provides a "shell" with which you can pass some instructions to pandas on how to aggregate chunks of values together.

So, you can sort of think of `.rolling()` as a way to tell pandas, OK, let's group together chunks of numbers in my data together and do something with them. 

Put another way, for each value you tell pandas to "roll", it will collect the previous N values and do something (something you tell it to do!) with the chunks altogether.

Often times we'll take a rolling mean of the last 30 days worth of data, for example, which is the mean of the last 30 values in the series. (You can also use `center=True`, which uses 15 values in the future and 15 values in the past and is sometimes useful) 

````python
# You can't use `.rolling()` alone, we need to tell pandas what to do with the rolling chunks. 
df_by_date['likes'].rolling(30)
````

````python
# Let's do a rolling mean.
# You'll notice the first 30ish values are empty, since you can't take the average of nothing! 
df_by_date['2015':'2018']['likes'].rolling(30).mean().head()
````

````python
# Let's plot it with pandas convenient plotting function
df_by_date['2015':'2018']['likes'].rolling(30).mean().plot()
````

## Let's do a little investigation to answer the question: 'What types of posts work best for us?'

#### We can't simply grab the posts with the most likes, we have to 'normalize' a little bit since we're (hopefully) constantly growing our userbase. E.G. You can't compare September 2017 likes-per-post to January 2016 likes-per-post because we have way more followers than then (and thus more likes per post). 

````python
# Let's make a little helper function
def show_post(df, month):
    print("Most liked post for the month of", month)
    print("----------------")
    url = str(df[month].sort_values('likes', ascending=False)['urls'][0][0])
    print("Caption:", str(df[month].sort_values('likes', ascending=False)['caption'][0]))
    print("----------------")
    print("")
    return url
````

````python
url = show_post(df_by_date, 'June 2017')
Image(url,width=300, height=300)
````

````python
url = show_post(df_by_date, 'October 2017')
Image(url,width=300, height=300)
````

````python
url = show_post(df_by_date, 'February 2018')
Image(url,width=300, height=300)
````

### One of the best ways to see how engaged your followers are is the ratio of likes to followers. The smaller the ratio the better (it means your followers tend to like your content and are willing to "double tap that post", otherwise they are probably just bored by your feed and scrolling past!)

_Let's look at the mean likes-per-post of the last couple of months only_

_For this number, it's ok for it to be the most recent months only_

````python
df_by_date['April 2018':]['likes'].mean()
````

_Let's store that in a variable to do some math with. Again, we could do this all at once but it's so much better to type a little more for your future self and anyone else that may look at your work!_

````python
avg_last_two_months_likes = df_by_date['April 2018':]['likes'].mean()
````

````python
# Based on code by William Feirie # 
import requests
import re

username_extract = 'messengercoffee'
url = 'https://www.instagram.com/'+ username_extract
r = requests.get(url)
m = re.search(r'"edge_followed_by":\{"count":([0-9]+)\}', str(r.content))
total_followers = int(m.group(1))
````

````python
total_followers
````

````python
print("Ratio of likes-per-post to followers is: {} to {}".format(1, int(total_followers / avg_last_two_months_likes)))
````
